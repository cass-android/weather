```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = requests.get(r'http://climate.weather.gc.ca/historical_data/search_historic_data_stations_e.html?searchType=stnProv&timeframe=1&lstProvince=&optLimit=yearRange&StartYear=1840&EndYear=2019&Year=2019&Month=5&Day=12&selRowPerPage=25')
```

```python
print(url.text)
```

```python
soup = BeautifulSoup(url.content, 'lxml')
```

```python
raw_stations = soup.find_all('input', {'name': 'StationID'})
```
